## Merge parctice 

```python
a_df = pd.DataFrame(
    {
        "Country": ["Germany", "France", "Belgium", "Finland"],
        "Population": [82.8, 67.2, 11.4, 5.5],
        "Capital" : ["Berlin", "Paris", "Brussels", "Helsinki"]
    }
)
```

```python
a_df.index = range(1, len(a_df) + 1)
```

```python
a_df
```

```python
b_df = pd.DataFrame(
    {
        "Country": ["Germany", "France", "Belgium", "Canada"],
        "HDI" : [0.936, 0.901, 0.916, 0.926]
    }
)
```

```python
b_df.index = range(1, len(b_df) + 1)
```

```python
b_df
```

```python
inner_merged_df = pd.merge(a_df,b_df,on="Country")
inner_merged_df.index = range(1, len(inner_merged_df) + 1)
inner_merged_df
```

### Left merge practice

```python
left_merged_df = pd.merge(a_df, b_df,on="Country", how="left")
left_merged_df.index = range(1, len(left_merged_df) + 1)
left_merged_df
```

### Right merge practice

```python
right_merged_df = pd.merge(a_df,b_df, on = "Country", how="right")
right_merged_df.index= range(1, len(right_merged_df) + 1)
right_merged_df
```

### Check your code

```python
outer_merged_df = pd.merge(a_df, b_df, on="Country", how="outer")
outer_merged_df.index = range(1, len(outer_merged_df) + 1)
outer_merged_df
```

### Country 

```python
countries_df = pd.read_csv("data/dictionary.csv")
countries_df.index = range(1, len(countries_df) + 1)
countries_df
```

### Summer Games

```python
summer_df = pd.read_csv("data/summer.csv")
summer_df = summer_df.rename(columns={"Country": "Code"})
summer_df.index = range(1, len(summer_df) + 1)
summer_df
```

### Winter Games

```python
winter_df = pd.read_csv("data/winter.csv")
winter_df = winter_df.rename(columns={"Country": "Code"})
winter_df.index = range(1, len(winter_df) + 1)
winter_df
```

### Combining The Data

```python
summer_countries_df = pd.merge(countries_df, summer_df, on="Code")
summer_countries_df.index = range(1, len(summer_countries_df) + 1)
summer_countries_df["Season"] = "Summer"
summer_countries_df
```

```python
winter_countries_df = pd.merge(countries_df, winter_df, on="Code")
winter_countries_df.index = range(1, len(winter_countries_df) + 1)
winter_countries_df["Season"] = "Winter"
winter_countries_df
```

```python
all_df = pd.concat([summer_countries_df, winter_countries_df])
all_df
```

### Top Countries Analysis

```python
filtered_df = all_df[all_df['Year'] >= 1984]
filtered_df
```

```python
grouped_df = filtered_df.groupby("Country")["Medal"].count()
grouped_df
```

```python
top_10_df = grouped_df.nlargest(10).reset_index()
top_10_df.columns = ['Country', 'Medal Count']
top_10_df.plot(kind="bar", x='Country', y='Medal Count').grid(True)
top_10_df
```

## Top 10 Countries (by Season) since 1984

```python
grouped_s_w_df = filtered_df.groupby(['Country', 'Season'])['Medal'].count()
grouped_s_w_df
```

```python
unstacked_df = grouped_s_w_df.unstack()
unstacked_df['Total'] = unstacked_df['Summer'] + unstacked_df['Winter']
unstacked_df
```

```python
top_10_season_df = unstacked_df.nlargest(10, 'Total')[['Summer', 'Winter', 'Total']]
top_10_season_df[['Summer', 'Winter']].plot(kind="bar").grid(True)
top_10_season_df
```
