# U.S. Oil and Gas Production Analysis

Analyzing the [Kaggle Dataset](https://www.kaggle.com/djzurawski/us-oil-and-gas-production-june-2008-to-june-2018) with information about Oil and Gas production in the US from June 2008 to June 2018.

## Loading Modules

```python
import numpy as np
import pandas as pd
import matplotlib
```

---

Let's load the Gas production:


```python
file = "data/U.S._natural_gas_production.csv"
gas_df = pd.read_csv(file, decimal=",")
gas_df.head(3)
```

```python
# Check how many rows and columns in the dataframe
gas_df.shape
```

```python
# Display all the available column names in the dataframe
gas_df.columns
```

```python
# Display additional info about each columns such as data types and number of non-null values
gas_df.info()
```

## Converting the Month column to datetime

```python
gas_df['Month'] = pd.to_datetime(gas_df['Month'])
```

```python
month_type = gas_df['Month'].dtype
```

```python
gas_df['Month'].dt.year.head()
```

```python
gas_df['Month'].dt.month.tail()
```

---

## Yearly Gas production


```python
gas_df['Month'] = pd.to_datetime(gas_df['Month'])
gas_df.set_index('Month', inplace=True)
yearly_gas_df = gas_df.groupby(gas_df.index.year).sum()
yearly_gas_df.head()
```

## Plotting results

```python
plot = yearly_gas_df.filter(items=["U.S."]).plot(kind="bar")
plot.grid(True)
```

```python
plot.set_xlabel("Year")
```

## Discarding rows with Boolean Indexing

```python
filtered_yearly_gas_df = yearly_gas_df[np.logical_and(yearly_gas_df.index > 2008,yearly_gas_df.index < 2018)]
filtered_yearly_gas_df
```

## State production

Let's have a look at the yearly production of some specific states

```python
plot = filtered_yearly_gas_df.filter(items=['Colorado', 'Louisiana', 'Ohio', 'Utah']).plot()
plot.set_xlabel("Year")
plot.grid(True);
```

## Comparing with Crude Oil Production


---

Let's load the Oil production:

```python
file = "data/U.S._crude_oil_production.csv"
oil_df= pd.read_csv(file, decimal=",")
oil_df.head(3)
```

```python
# Check how many rows and columns in the dataframe
gas_df.shape
```

```python
# Display all the available column names in the dataframe
gas_df.columns
```

```python
# Display additional info about each columns such as data types and number of non-null values
gas_df.info()
```

## Converting the Month column to datetime

```python
oil_df["Month"] = pd.to_datetime(oil_df["Month"])
month_type = oil_df['Month'].dtype
```

---

## Yearly Oil Production

```python
oil_df['Month'] = pd.to_datetime(oil_df['Month'])
oil_df.set_index('Month', inplace=True)
yearly_oil_df = oil_df.groupby(oil_df.index.year).sum()
yearly_oil_df.head()
```

```python
yearly_oil_df.columns = yearly_oil_df.columns.str.strip()
yearly_oil_df.columns
```

```python
crude_plot = yearly_oil_df.filter(items=['U.S. Crude Oil']).plot(kind='bar')
crude_plot.set_xlabel("Year")
crude_plot.grid(True);
```

## Discarding rows with Boolean Indexing

```python
filtered_yearly_oil_df = yearly_oil_df[np.logical_and(yearly_oil_df.index > 2008,yearly_oil_df.index < 2018)]
filtered_yearly_oil_df
```

### Merging Oil and Gas Production

```python
total_gas = filtered_yearly_gas_df[["U.S."]].rename(columns={"U.S.": "Gas"})
total_oil = filtered_yearly_oil_df[["U.S. Crude Oil"]].rename(columns={"U.S. Crude Oil": "Crude Oil"})
```

```python
merged_df = pd.concat([total_oil,total_gas], axis=1)
merged_df
```

```python
merged_plot = merged_df.plot(kind="bar")
merged_plot.set_xlabel("Year")
merged_plot.legend(["Gas (Millions of Cubic feet) ", "Crude Oil (Thousands of barrels)"])
merged_plot.grid(True)
```

```python
total_gas = filtered_yearly_gas_df[["U.S."]].rename(columns={"U.S.": "Gas"})
total_oil = filtered_yearly_oil_df[["U.S. Crude Oil"]].rename(columns={"U.S. Crude Oil": "Crude Oil"})
```

```python
merged_df_1 = pd.concat([total_oil,total_gas],axis="columns")
merged_df_2 = pd.concat([total_gas,total_oil],axis="columns")
```
