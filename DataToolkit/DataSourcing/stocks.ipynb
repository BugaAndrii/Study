### Import modules


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import requests
```

### Apple

```python
apple = "https://iex.lewagon.com/stable/stock/aapl/chart/3m"
apple_data = requests.get(apple).json()
apple_data
```

```python
apple_df = pd.DataFrame(apple_data)
apple_df
```

```python
apple_df["date"] = pd.to_datetime(apple_df["date"],format="%Y-%m-%d")
apple_df = apple_df.set_index("date")
```

```python
apple_df
```

```python
apple_df["close"].plot().grid(True);
```

```python
apple_df[["open", "close", "high", "low"]].plot().grid(True)
```

```python
apple_df[["open", "close", "high", "low"]].plot(figsize=(12,4)).grid(True)
```

### Google

```python
google = "https://iex.lewagon.com/stable/stock/goog/chart/3m"
google_data = requests.get(google).json()
google_data
google_df = pd.DataFrame(google_data)
google_df
google_df["date"] = pd.to_datetime(google_df["date"],format="%Y-%m-%d")
google_df = google_df.set_index("date")
google_df
```

```python
apple_df
```

```python
all_df = pd.concat([google_df["close"], facebook_df["close"], amazon_df["close"], apple_df["close"]], axis=1)
all_df.columns = ["google", "facebook", "amazon", "apple"]
all_df
```

```python
all_df_normalaized = all_df / all_df.iloc[0]
all_df_normalaized
```

```python
all_df_normalaized.plot(figsize=(12,6)).grid(True)
```
